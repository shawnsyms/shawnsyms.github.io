#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

````python
def stddevcalc(column):
    n = len(column)
    mu = np.mean(column)
    sumofsquares = 0
    for i in column:
        sumofsquares = sumofsquares + (i-mu)**2
    stddev = (1/n * sumofsquares)**(1/2)
    return stddev
````
